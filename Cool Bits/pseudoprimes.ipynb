# Pseudoprimes

Pseudoprimes (at least in the context of this codebase) are prime-like numbers whose main attribute is, that they get more sparse the larger they are. Pseudoprimes are generated in a geometric progression and are visualized in a plot below.

```python
# Imports
import matplotlib.pyplot as plt

# Importing the supportive functions from additional files
from Utilities.input_gen_utils import *  # Input generation
```

```python
# ---
# Plotting pseudoprimes

pseudoprimes = generate_pseudoprimes_in_range(1, 10000000)
print(f'Finished generating pseudoprimes list with {len(pseudoprimes)} examples.')


def get_subset_of_primes(output_size, primes_list):
    step = int(len(primes_list) / output_size)
    print(f'Step size is {step}; total primes: {len(primes_list)}')

    output = []
    i = 0
    while i < len(primes_list):
        output.append(primes_list[i])
        i += step

    return output  


# Plotting
plt.figure(figsize=(10, 6))
plt.plot(pseudoprimes, range(len(pseudoprimes)), 'bo-', markersize=0.01, label='Pseudoprimes')
plt.xlabel('Pseudoprime Value')
plt.ylabel('Index')
plt.title('Pseudoprimes Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
